```python
import numpy as np
import pandas as pd
import datetime
import os
from scipy import stats
from collections import defaultdict
#import statsmodels.formula.api as sm
```

```python
with open('DJticker.csv') as file:
    tickers=file.read().split(',') 
```

```python
price = defaultdict(str)
sentiment=defaultdict(str)
dt=defaultdict(str)
res=defaultdict(str)
```

```python
for ticker in tickers:
    pricepFile = ticker+"price.csv"
    sentimentFile = ticker+"SentimentRelAll.csv"
    priceData = pd.read_csv(pricepFile, sep=",", skiprows=1,
                     names = ["Date","Ticker","Price"], low_memory=False)
    priceData.Date = priceData.Date.apply(lambda x: pd.to_datetime(x,format='%m/%d/%Y'))
    sentimentData = pd.read_csv(sentimentFile, sep=",", skiprows=1,
                     names = ["UTCTime","Ticker","Sentiment","Relevance","MarketDate"], low_memory=False)
    sentimentData.MarketDate = sentimentData.MarketDate.apply(lambda x: pd.to_datetime(x,format='%Y-%m-%d'))
    price[ticker]=priceData#
    sentiment[ticker]=sentimentData#
    dt[ticker] = sentiment[ticker].groupby(["MarketDate"])["Sentiment"].sum()
    res[ticker] = pd.merge(price[ticker], dt[ticker].to_frame(), left_on = 'Date', right_index = True)
    res[ticker]["Return"] = np.log(res[ticker].Price).diff()
    res[ticker]["Mark"] = res[ticker].Sentiment.apply(lambda x: 1 if x>0 else (-1 if x<0 else 0))
    res[ticker]["LagReturn"] = res[ticker]["Return"].shift(-3)
```

```python
for ticker in tickers:
    print(ticker)
    res[ticker].groupby(['Mark']).agg({'LagReturn': ['sum', 'count']})
    
```

```python
result = sm.ols(formula="Mark ~ Sentiment", data=res).fit()
```
